## Checking for correlation

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
nycab=pd.read_excel(r'Desktop\Updated Neighbourhoods.xlsx')
```

```python
nycab.describe()
```

```python
nycab.head()
```

```python
df1=nycab.drop(['neighbourhood_group_cleansed','NeighbourhoodArea','Listings'], axis=1)
df1.head()
```

```python
df1.describe()
```

```python
correlation = df1.corr()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(correlation,cbar = True, square = True, cmap= 'Blues', annot = True, fmt = '.1f')
```

##### No strong correlation found overall

### Creating a function to display for all neighbourhoods

```python
def correl(name):
    dfnbh=nycab.loc[(nycab['neighbourhood_group_cleansed']==name)]
    #dfnbh.tail()
    df2=dfnbh.drop(['neighbourhood_group_cleansed','NeighbourhoodArea','Listings'], axis=1)
    correlation_nbh = df2.corr()
    plt.figure(figsize=(10,10))
    sns.heatmap(correlation_nbh,cbar = True, square = True, cmap= 'Blues', annot = True, fmt = '.1f')
```

```python
correl('Bronx')
```
